```python
import re
import json

def parse_srt_file(file_path):
    with open(file_path, 'r') as file:
        srt_data = file.read()
    
    # Split the SRT data into individual subtitle blocks
    subtitle_blocks = srt_data.strip().split('\n\n')
    
    # Initialize a list to store the parsed subtitles
    subtitles = []

    # Iterate over each subtitle block
    for block in subtitle_blocks:
        # Extract the subtitle lines
        lines = block.strip().split('\n')
        
        # Extract the subtitle sequence number
        sequence_number = int(lines[0])
        
        # Extract the subtitle timestamps
        timestamps = re.findall(r'(\d{2}:\d{2}:\d{2},\d{3})', lines[1])
        start_time, end_time = timestamps
        
        # Extract the subtitle text lines
        text_lines = lines[2:]
        
        # Join the text lines to form the subtitle text
        text = '\n'.join(text_lines)
        
        # Create a subtitle dictionary
        subtitle = {
            'sequenceNumber': sequence_number,
            'startTime': start_time,
            'endTime': end_time,
            'text': text
        }
        
        # Append the subtitle dictionary to the subtitles list
        subtitles.append(subtitle)
    
    # Return the parsed subtitles as JSON
    return json.dumps(subtitles, indent=4)

# Example usage
srt_file_path = '2/in-video.srt'
json_subtitles = parse_srt_file(srt_file_path)
print(json_subtitles)
```
